##### dropping correlated feature

```python
# By using this function we can select correlated features
# it will remove the first feature that is correlated with anything other feature
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
col_corr = corr_features = correlation(df, 0.8)
print(col_corr)
```

## hyperparamter tuning (using gridSearch)

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_fscore_support

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import CategoricalNB

from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
```

```python
## gridsearch will try possible parameters and return the best
from sklearn.model_selection import GridSearchCV
```

### defining models & hyperparameters 

```python
# Random Forest
param_grid_rf = {
    'max_depth': np.arange(1,20),
    'n_estimators': np.arange(1,10),
    'criterion' : ["gini", "entropy", "log_loss"]
    # 'max_leaf_nodes': np.arange(1,51,5),
}


# SVC
mrange_svc = np.arange(0.01, 0.11, 0.01)
param_grid_svc = {
    'C': mrange_svc,
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
}


# Decision Tree
param_grid_dt = {
    'max_depth': np.arange(1,20),
    'criterion' : ["gini", "entropy", "log_loss"]
}



# Logistic Regression
mrange_log = np.arange(0.1, 1.1, 0.1)
param_grid_log = {'C': mrange_log}



# Naive Bais
mrange_nb = np.arange(0.1, 1.1, 0.1)
param_grid_nb = {
    'alpha': mrange_nb
}



# KNN
mrange_knn = np.arange(1,100,2)
param_grid_knn = {
    'n_neighbors': mrange_knn,
}
```

### fitting (function)

```python
def grid_fit(clf, params, n_jobs:int = 1):
    grid_clf = GridSearchCV(clf, params, cv=5, verbose=3, return_train_score=True, n_jobs=n_jobs)
    grid_clf.fit(X_train, y_train)
    return grid_clf
```

### evalaute (function)

```python
fscore_0 = {}
fscore_1 = {}
fscore_2 = {}
fscore_3 = {}
def evalaute(clf):
    global fscore_0, fscore_1, fscore_2, fscore_3


    # best params
    if(clf.__class__.__name__ == 'GridSearchCV'):
        print(f"Best Parameters: {clf.best_params_} with best score {clf.best_score_}")
        print()
        print()
      



    # predecting
    y_pred_test = clf.predict(X_test)
    y_pred_train = clf.predict(X_train)




    # evaluating train
    print('evalauting train data')
    print(f'train accuracy: {accuracy_score(y_train,y_pred_train)*100}')
    print(classification_report(y_train,y_pred_train))
    cm_display = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_train,y_pred_train)
                                        , display_labels=clf.classes_)
    cm_display.plot()
    plt.show()




    # evaluating test
    print('evalauting test data')
    print(f'test accuracy: {accuracy_score(y_test,y_pred_test)*100}')
    print(classification_report(y_test,y_pred_test))
    cm_display = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,y_pred_test)
                                        , display_labels=clf.classes_)
    cm_display.plot()
    plt.show()




    # saving f1 score for each classifier    
    report = precision_recall_fscore_support(y_test,y_pred_test)
    if(clf.__class__.__name__ == 'GridSearchCV'):
        clf_name=clf.estimator.__class__.__name__
    else:
        clf_name=clf.__class__.__name__
        
    fscore_0.update({clf_name: report[2][0]})
    fscore_1.update({clf_name: report[2][1]})
    fscore_2.update({clf_name: report[2][2]})
    fscore_3.update({clf_name: report[2][3]})
```

### building models

#### decision tree 

```python

dt_grid = grid_fit(DecisionTreeClassifier(random_state=42), param_grid_dt, -1)
evalaute(dt_grid)
```

#### Random Forest


```python
rf_grid = grid_fit(RandomForestClassifier(random_state=42), param_grid_rf, -1)
evalaute(rf_grid)
```

#### Logistic Regression


```python
log_grid = grid_fit(LogisticRegression(random_state=42), param_grid_log, -1)
evalaute(log_grid)
```

#### Naive bayes


```python
# nb_grid = grid_fit(CategoricalNB(), param_grid_nb, -1)
# evalaute(nb_grid)
```

#### SVC

```python
svc_grid = grid_fit(SVC(random_state=42), param_grid_svc, -1)
evalaute(svc_grid)
```

#### KNN

```python
knn_grid = grid_fit(KNeighborsClassifier(), param_grid_knn, -1)
evalaute(knn_grid)
```

```python
acc = knn_grid.cv_results_['mean_test_score']
err_rate = 1 - acc

plt.figure(figsize = (15,6))
plt.plot(range(1,100,2),err_rate,color = 'black',linestyle = '--',marker = 'o',markerfacecolor='red',markersize = 8)
plt.title('Error Rate vs K')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.yticks(np.arange(min(err_rate), max(err_rate),0.01))
plt.xticks(np.arange(1, 100,2))
plt.show()
```

### Best model

#### plotting fscores

```python
fscores = fscore_0,fscore_1,fscore_2,fscore_3
```

```python
import pprint
for i,fscore in enumerate(fscores):
    print(f'fscore_{i}')
    pprint.pprint(fscore)
    print()
```

```python
plt.figure(figsize=(30,20))
for i, fscore in enumerate(fscores):
    plt.subplot(2, 2, i + 1)

    keys = np.array(list(fscore.keys()))
    vals = np.array(list(fscore.values()))
    
    plt.xlabel("Models")
    plt.ylabel(f'fscore_{i}')
    # plt.yticks(np.arange(0, max(vals), max(vals)/15))
    plt.ylim(bottom=0 , top=1 )
    plt.bar(keys, vals)

plt.show()
```

#### Ploting best model

```python
mdls_val_grid = {}
for clf in (dt_grid, rf_grid, log_grid, knn_grid, svc_grid):
            mdls_val_grid[clf.estimator.__class__.__name__] =  clf.best_score_
    
```

```python
## sorting models
sorted_mdls_val = dict(sorted(mdls_val_grid.items(), key=lambda x: x[1], reverse=True))
sorted_mdls_val
```
